# Task 1

In this task, I get all the excel data of task 1, turn it into python lists and plot counts vs beta for the 4 cases.
I also write define a peak(data) function which gets the 2 peaks of the curves

Equations used:

$n\lambda = 2d\sin(\beta)$

```python
#Getting all datasets from Excel and converting them into lists. Dumb me forgot that all the betas are the same, but 
#this mistake will come in handy in the smallPeak(data) function

a = pd.read_excel('rawdata.xlsx', usecols='A')
b = pd.read_excel('rawdata.xlsx', usecols='B')
c = pd.read_excel('rawdata.xlsx', usecols='C')
d = pd.read_excel('rawdata.xlsx', usecols='D')
e = pd.read_excel('rawdata.xlsx', usecols='E')
f = pd.read_excel('rawdata.xlsx', usecols='F')
g = pd.read_excel('rawdata.xlsx', usecols='G')
h = pd.read_excel('rawdata.xlsx', usecols='H')

betaOne = [i for i in a['Table 1']]
countOne = [i for i in b['Unnamed: 1']]
betaTwo = [i for i in c['Unnamed: 2']]
countTwo = [i for i in d['Unnamed: 3']]
betaThree = [i for i in e['Unnamed: 4']]
countThree = [i for i in f['Unnamed: 5']]
betaFour = [i for i in g['Unnamed: 6']]
countFour = [i for i in h['Unnamed: 7']]


#Function to return value of the small and large peak 

def bigPeak(data):
    dummy = [i for i in data]
    dummy.sort()
    a = dummy[-1]
    b = data.index(a)
    c = betaFour[b]
    return c

def smallPeak(data):
    dummy_new = [i for i in data]
    a = bigPeak(data)
    index_a = betaOne.index(a)
    dummy_new = dummy_new[0:index_a - 2]
    return bigPeak(dummy_new)

def peak(data):
    return [smallPeak(data), bigPeak(data)]

```

```python
#Plot 1


import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
plt.title('Count Rate vs Angle with following confirguration:\n U = 35kV, I = 1.0mA, $\Delta t = 4s,  \Delta $' + r'$ \beta = 0.1^{\circ},  \beta_{min} = 2.5^{\circ},$ and $\beta_{max} = 12.5^{\circ}$')
plt.xlabel(r'$ \beta (^{\circ})$')
plt.ylabel(r'$ Count Rate (s^{-1})$')
plt.plot(betaOne, countOne, color='red', label='Count Rate vs angle')
plt.legend()
plt.show()

print("Peaks are " + str(peak(countOne)))



```

# Task 3

New dataset. I will use this dataset to plot the $\frac{1}{U}$ vs. $\lambda_{min}$ and will use  numpy.polyfit(x,y,1) (LINEST equivelent) to derive planck's constant ($\frac{slope * e}{c}$, where $e$ and $c$ are known constants)

Equation used:

$\lambda_{min} = \frac{hc}{eU}$

So the slope of $\lambda_{min}$ vs $1/U$ is $A = hc/e$

ie. 

$h = Ae/c$

```python
#Importing New Datasets

a = pd.read_excel('rawdata2.xlsx', usecols='A')
b = pd.read_excel('rawdata2.xlsx', usecols='B')
c = pd.read_excel('rawdata2.xlsx', usecols='C')
d = pd.read_excel('rawdata2.xlsx', usecols='D')
e = pd.read_excel('rawdata2.xlsx', usecols='E')
f = pd.read_excel('rawdata2.xlsx', usecols='F')
g = pd.read_excel('rawdata2.xlsx', usecols='G')
h = pd.read_excel('rawdata2.xlsx', usecols='H')

lambdaOne = [i for i in a['Table 1']]
countOne = [i for i in b['Unnamed: 1']]
lambdaTwo = [i for i in c['Unnamed: 2']]
countTwo = [i for i in d['Unnamed: 3']]
lambdaThree = [i for i in e['Unnamed: 4']]
countThree = [i for i in f['Unnamed: 5']]
lambdaFour = [i for i in g['Unnamed: 6']]
countFour = [i for i in h['Unnamed: 7']]
```

```python
y_err1 = sem(countOne)
y_err2 = sem(countTwo)
y_err3 = sem(countThree)
y_err4 = sem(countFour)

plt.errorbar(lambdaOne, countOne,yerr=y_err1, label = 'U = 22kV, Error = ' + str(round(y_err1,2)))
plt.errorbar(lambdaTwo, countTwo,yerr=y_err2, label = 'U = 26kV, Error = '+ str(round(y_err2,2)))
plt.errorbar(lambdaThree, countThree,yerr=y_err3, label = 'U = 30kV, Error = '+ str(round(y_err3,2)))
plt.errorbar(lambdaFour, countFour,yerr=y_err4, label = 'U = 34kV, Error = '+ str(round(y_err4,2)))
plt.legend(loc='upper left')
plt.xlabel('$\lambda$ (pm)')
plt.ylabel('Counts ($s^{-1}$)')
plt.title('Counts vs. Wavelength for Obtaining $\lambda_{min}$')

```

```python
from uncertainties import ufloat
x_int = lambda m,b: -b/m

#LINEST equivelent
m1,b1 = np.polyfit(lambdaOne[30:40],countOne[30:40], 1)
m2,b2 = np.polyfit(lambdaTwo[17:34],countTwo[17:34], 1)
m3,b3 = np.polyfit(lambdaThree[13:20],countThree[13:20], 1)
m4,b4 = np.polyfit(lambdaFour[9:16],countFour[9:16], 1)

#lambdaMin and 1/U (with errors)

waveMin = [x_int(m4,b4),x_int(m3,b3),x_int(m2,b2),x_int(m1,b1)]
waveMinErr = sem(waveMin)


uRec = [1/34,1/30,1/26,1/22]
uRecErr = sem(uRec)


A = ufloat(np.polyfit(uRec, waveMin,1)[0]/1000000000, waveMinErr/1000000000)
c = 299792458
e = 1.60217662e-19
planckFound = A*e/c
print('Planck Found: ' + str(planckFound))
```
